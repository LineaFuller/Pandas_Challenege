```python
# Import Dependencies 
import pandas as pd
```

```python
# Path to file 
file_path = "Resources/purchase_data.csv"
```

```python
# Read Purchase data and store into Pandas df
purchase_df = pd.read_csv(file_path)
purchase_df.head()
```

```python
# Player count
player_count = purchase_df["SN"].nunique()
total_players = purchase_df["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

```python
# Purchasing Analysis
unique_items = purchase_df["Item ID"].nunique()
average_price = round(purchase_df["Price"].mean(),2)
num_purchase = purchase_df.shape[0]
total_revenue = purchase_df["Price"].sum()

purchasing_Analysis_df = pd.DataFrame({
    "Number of Unique Items": [unique_items],
    "Average Price": [average_price],
    "Number of Purchase": [num_purchase],
    "Total Revenue": [total_revenue]
})

purchasing_Analysis_df["Average Price"]=purchasing_Analysis_df["Average Price"].map("${:}".format)
purchasing_Analysis_df["Total Revenue"]=purchasing_Analysis_df["Total Revenue"].map("${:,.2f}".format)

purchasing_Analysis_df

```

```python
# Gender Demographics

# Assign Values
player_df = purchase_df[["SN", "Age", "Gender"]].drop_duplicates()
gender_gb = player_df.groupby("Gender")
gender_count = gender_gb["SN"].count()
percentage = round(gender_count/total_players *100, 2)
percentage

# Create DataFrame
gender_df = pd.DataFrame({
    "Total Count": gender_count,
    "Percentage of Players": percentage
})

gender_df["Percentage of Players"]=gender_df["Percentage of Players"].map("{:}%".format)
gender_df
```

```python
# Purchasing Analysis
price_analysis_df = purchase_df.groupby("Gender")

gender_purchase_count = price_analysis_df['Purchase ID'].count()
average_price = round(price_analysis_df["Price"].mean(), 2)
total_purchase_value = round(price_analysis_df["Price"].sum(), 2)
average_total_purchase_per_gender = round(total_purchase_value/gender_count, 2)


gender_purchase_analysis_df = pd.DataFrame({
    "Purchase Count": gender_purchase_count,
    "Average Purchase Price": average_price,
    "Total Purchase Value": total_purchase_value,
    "Average Total Purchase Per Person": average_total_purchase_per_gender
})

gender_purchase_analysis_df["Average Purchase Price"] = gender_purchase_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase_analysis_df["Total Purchase Value"] = gender_purchase_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_analysis_df["Average Total Purchase Per Person"] = gender_purchase_analysis_df["Average Total Purchase Per Person"].map("${:,.2f}".format)
gender_purchase_analysis_df.head()


# gender_purchase_analysis_df = pd.DataFrame()
# gender_purchase_analysis_df["Purchase Count"]= gender_purchase_count['SN']
# gender_purchase_analysis_df["Average Purchase Price"]=
# gender_purchase_analysis_df["Total Purchase Value"]=
# gender_purchase_analysis_df["Average Total Purchase Per Person"]=
```
